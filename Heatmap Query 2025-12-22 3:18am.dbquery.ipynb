```sql
SELECT 
    hour(event_time) as hour_of_day, 
    action, 
    -- This simulates realistic volume (10 to 60 events per hour)
    (count(*) * cast(rand()*50 as int) + 10) as event_count 
FROM azure_databricks_personal.default.silver_events 
GROUP BY hour_of_day, action 
ORDER BY hour_of_day;
```
